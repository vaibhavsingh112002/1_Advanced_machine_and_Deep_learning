```python
import tensorflow as tf
```

initializing the input sensor

```python
a = tf.constant([7,8,13,11],dtype = tf.float64)
b = tf.constant([2,13,14,5],dtype = tf.float64)
```

printing the input sensor

```python
print('a',a)
print('b',b)
```

finding truth value

```python
res = tf.math.less(x=a,y=b)
```

printing results

```python
print('Result:',res)
```
